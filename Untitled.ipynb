```python
csse_confirmed_df = pd.read_csv(csse_confirmed_path)

hubei = csse_confirmed_df[(csse_confirmed_df['Country/Region'] == 'China' ) & (csse_confirmed_df['Province/State'] == 'Hubei' )]
hubei['Country/Region'] = 'China - Hubei'
csse_confirmed_df = pd.concat([csse_confirmed_df, hubei])

csse_deaths_df = pd.read_csv(csse_deaths_path)

hubei = csse_deaths_df[(csse_deaths_df['Country/Region'] == 'China' ) & (csse_deaths_df['Province/State'] == 'Hubei' )]
hubei['Country/Region'] = 'China - Hubei'
csse_deaths_df = pd.concat([csse_deaths_df, hubei])

csse_recovered_df = pd.read_csv(csse_recovered_path)

hubei = csse_recovered_df[(csse_recovered_df['Country/Region'] == 'China' ) & (csse_recovered_df['Province/State'] == 'Hubei' )]
hubei['Country/Region'] = 'China - Hubei'
csse_recovered_df = pd.concat([csse_recovered_df, hubei])

csse = {'confirmed':csse_confirmed_df,'deaths':csse_deaths_df,'recovered':csse_recovered_df}
```

```python
def transpose_df(df):
    cols = [x for x in df.columns if x not in ['Country/Region','Province/State','Lat', 'Long']]
    df = df.groupby(['Country/Region'])[cols].agg('sum').reset_index().set_index('Country/Region').T
    df.columns.names = ['']
    df.reset_index(inplace=True)
    df.rename(columns={'index':'Date'}, inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    return df
```

```python
def delta(row, df, country, firstday = 0, days = 1):
    if row.dayofyear - days <= firstday:
        return 0
    else:
        return int(row[country] - df[df.dayofyear == (row.dayofyear - days)][country])
```

```python
checkpoint = 100
```

```python
def dayplus(row, df, country):
    first_day = df[df[country] > checkpoint].dayofyear.min()
    if np.isnan(first_day) or row.dayofyear <= first_day:
        return 0
    else:
        return row.dayofyear - first_day
```

```python
def create_features(df):
    countries = [x for x in df.columns if x not in ['Date']]
    not_china_cols = [x for x in countries if x not in ['China']]
    df['non-China'] = df[not_china_cols].sum(axis=1)
    not_china_italy_cols = [x for x in countries if x not in ['China', 'Italy']]
    df['non-China-Italy'] = df[not_china_italy_cols].sum(axis=1)
    df['world'] = df[countries].sum(axis=1)
    df['dayofyear'] = df.apply(lambda x : x.name.dayofyear, axis = 1)
    for country in tqdm(countries + ['non-China', 'world', 'non-China-Italy']):
        df[str(country) + '_daily'] = df.apply(lambda x : delta(x, df, country, firstday = df.dayofyear.min()), axis = 1)
        df[str(country) + '_daysplus'] = df.apply(lambda x : dayplus(x, df, country), axis = 1)
    return df
```

```python
for key, value in csse.items():
    csse[key] = create_features(value)
```

```python
def pop_perc(df):
    for col in tqdm(df.columns):
        if not col.endswith('_daysplus') and col not in ['Date', 'dayofyear'] + missing:
            #print(col)
            #print(int(population_df[population_df.Territory == col.split('_')[0]]['Population']))
            #display(df[col])#/(int(population_df[population_df.Territory == col.split('_')[0]]['Population'])))
            df[col] = df[col]/(int(population_df[population_df.Territory == col.split('_')[0]]['Population']))
    return df
```

```python
def dayplus_perc(row, df, country):
    first_day = df[df[country] > 0.00001].dayofyear.min()
    if np.isnan(first_day) or row.dayofyear <= first_day:
        return 0
    else:
        return row.dayofyear - first_day
```

```python
X = csse['confirmed']['Argentina'].values
size = int(len(X) * 0.1)
train, test = X[0:size], X[size:len(X)]
history = [x for x in train]
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    history.append(obs)
    print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)
# plot

pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
